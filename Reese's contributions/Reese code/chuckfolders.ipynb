```python
import os
import shutil
from pathlib import Path

# === Settings ===
base_dir = Path(r"D:\vc-research\vc-research\Batch3")
num_folders = 38

# === Get all PDF files ===
pdf_files = sorted([f for f in base_dir.glob("*.pdf")])

# === Create subfolders ===
subfolders = []
for i in range(1, num_folders + 1):
    subfolder = base_dir / f"Batch3_{i}"
    subfolder.mkdir(exist_ok=True)
    subfolders.append(subfolder)

# === Copy PDFs evenly ===
for idx, pdf in enumerate(pdf_files):
    target_folder = subfolders[idx % num_folders]
    shutil.copy2(str(pdf), target_folder / pdf.name)  # copy2 preserves metadata

print(f"✅ Successfully copied {len(pdf_files)} PDFs into {num_folders} subfolders.")
```
